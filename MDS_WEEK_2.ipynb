## 1. Computation on NumPy Arrays

```python
#Creating Arrays & Matrix Operations
import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("Addition")
print(A+B)

print("Subtraction")
print(A-B)

print("Multiplication")
print(A*B)

print("Matrix Multiplication")
print(A@B)
```

```python
#Slicing (Extract Elements)
import numpy as np

a = np.array([[10,20,30],
              [40,50,60],
              [70,80,90]])

print(a)

print("First Row")
print(a[0])

print("Second Column")
print(a[:,1])

print("First Two Rows")
print(a[0:2])

print("Last Two Columns")
print(a[:,1:])
```

```python
#np.random.seed()
import numpy as np

np.random.seed(49)

print(np.random.rand(2,3))
```

```python
#np.random.rand()
import numpy as np

np.random.seed(49)

a = np.random.rand(3,3)

print(a)
```

```python
#np.random.randint()
import numpy as np

np.random.seed(49)

a = np.random.randint(1,20,(3,3))

print(a)
```

```python
#Concatenate
import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])

c=np.concatenate((a,b))

print(c)
```

```python
#Stack, hstack, vstack
import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])

print("Stack")
print(np.stack((a,b)))

print("Horizontal Stack")
print(np.hstack((a,b)))

print("Vertical Stack")
print(np.vstack((a,b)))
```

```python
#Split Array
import numpy as np

a=np.array([10,20,30,40,50,60])

x=np.split(a,3)

print(x)
```

```python
#Aggregate Functions
import numpy as np

a=np.array([[2,8,4],
            [9,1,6]])

print("Minimum =",np.min(a))
print("Maximum =",np.max(a))

print("Argmin =",np.argmin(a))
print("Argmax =",np.argmax(a))

print("Sum =",np.sum(a))
print("Mean =",np.mean(a))
```

```python
#Boolean Masking
import numpy as np

a=np.array([5,10,15,20,25])

print(a>15)

print(a[a>15])
```

```python
#Broadcasting
import numpy as np

a=np.array([[1,2,3],
            [4,5,6]])

print(a+10)
```

```python
#Universal Functions
import numpy as np

a=np.array([1,4,9,16])

print("Square Root")
print(np.sqrt(a))

print("Square")
print(np.square(a))

print("Exponential")
print(np.exp(a))

print("Sine")
print(np.sin(a))
```

## 2. numpy.linalg Functions

```python
#Determinant
import numpy as np

A=np.array([[2,1],
            [3,4]])

print(np.linalg.det(A))
```

```python
#Inverse
import numpy as np

A=np.array([[2,1],
            [3,4]])

print(np.linalg.inv(A))
```

```python
#Transpose
import numpy as np

A=np.array([[1,2],
            [3,4]])

print(A.T)
```

```python
#Rank
import numpy as np

A=np.array([[1,2],
            [2,4]])

print(np.linalg.matrix_rank(A))
```

```python
#Eigen Values and Eigen Vectors
import numpy as np

A=np.array([[2,1],
            [1,2]])

value,vector=np.linalg.eig(A)

print(value)
print(vector)
```

```python
#Solve Linear Equations
import numpy as np

A=np.array([[2,1],
            [1,-1]])

B=np.array([5,1])

print(np.linalg.solve(A,B))
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("image.png")

img_array = np.array(img)

print("Image Shape:", img_array.shape)
```

```python
# Original Image
plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Crop the Image Using NumPy Indexing
cropped = img_array[100:400, 150:500]

plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
#Flip the Image Horizontally
horizontal_flip = img_array[:, ::-1]

plt.imshow(horizontal_flip)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
#Flip the Image Vertically
vertical_flip = img_array[::-1, :]

plt.imshow(vertical_flip)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```
