### Comparing Python Lists and NumPy Arrays Performance

This section demonstrates the performance differences between standard Python lists and NumPy arrays for common operations, especially with large datasets.

As you can see from the output, NumPy arrays are significantly faster for numerical operations, especially with large datasets, due to their optimized C implementations. This is why NumPy is preferred for scientific computing and data analysis in Python.

### Memory Usage Comparison

Let's compare the memory footprint of Python lists versus NumPy arrays for storing numerical data.

As shown above, NumPy arrays are significantly more memory-efficient for numerical data than Python lists. This is because Python lists store actual Python objects (which have their own overhead) and only store references to these objects, while NumPy arrays store raw data in a contiguous block of memory.

```python
import sys
import numpy as np

# Number of elements for comparison
num_elements = 100000

print(f"Comparing memory usage for {num_elements} elements:")

# --- Python List Memory Usage ---
python_list = list(range(num_elements))
# sys.getsizeof() reports the size of the list structure itself,
# which holds references (pointers) to the actual Python integer objects.
# It does NOT include the memory taken by the individual integer objects.
list_structure_size_bytes = sys.getsizeof(python_list)

print(f"\nPython List (structure size for {num_elements} elements): {list_structure_size_bytes} bytes")
print(f"  (Each element in a Python list is a separate Python object, adding significant overhead.)")

# --- NumPy Array Memory Usage ---
# NumPy arrays store elements contiguously and efficiently.
# .nbytes gives the total bytes consumed by the array's data buffer itself.
numpy_array = np.arange(num_elements, dtype=np.int64) # Using int64 for a typical numerical array
numpy_array_data_size_bytes = numpy_array.nbytes

print(f"\nNumPy Array (data size for {num_elements} int64 elements): {numpy_array_data_size_bytes} bytes")
print(f"  (NumPy stores elements directly in a contiguous block of memory, which is memory efficient.)")
```

```python
import time

# Define a large number of elements for comparison
size = 10**6

# --- Python List Operations ---
print(f"Performing operations on Python lists with {size} elements...")

# Creating a Python list
start_time = time.time()
py_list = list(range(size))
end_time = time.time()
print(f"Time taken to create Python list: {end_time - start_time:.6f} seconds")

# Element-wise operation (e.g., squaring each element)
start_time = time.time()
py_list_squared = [x**2 for x in py_list]
end_time = time.time()
print(f"Time taken for element-wise squaring (Python list): {end_time - start_time:.6f} seconds")

# --- NumPy Array Operations ---
print(f"\nPerforming operations on NumPy arrays with {size} elements...")

# Creating a NumPy array from a Python list
start_time = time.time()
np_array = np.array(range(size))
end_time = time.time()
print(f"Time taken to create NumPy array: {end_time - start_time:.6f} seconds")

# Element-wise operation (e.g., squaring each element)
start_time = time.time()
np_array_squared = np_array**2
end_time = time.time()
print(f"Time taken for element-wise squaring (NumPy array): {end_time - start_time:.6f} seconds")
```
